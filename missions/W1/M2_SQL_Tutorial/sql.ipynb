```python
import sqlite3
conn = sqlite3.connect('w3schools.db')
cursor = conn.cursor()
```

```python
def run_select(query):
    """SELECT 전용 — 결과를 표처럼 확인"""
    cursor.execute(query)
    rows = cursor.fetchall()
    cols = [d[0] for d in cursor.description]   # 컬럼명
    print(cols)
    for row in rows:
        print(row)
    return rows

def run_execute(query):
    """INSERT/UPDATE/DELETE/CREATE 등 — 커밋까지"""
    cursor.execute(query)
    conn.commit()
    print(f"OK. rowcount={cursor.rowcount}")   # 영향받은 행 수
```

SQL Aggregate Functions
An aggregate function is a function that performs a calculation on a set of values, and returns a single value.

Aggregate functions are often used with the GROUP BY clause of the SELECT statement. The GROUP BY clause splits the result-set into groups of values and the aggregate function can be used to return a single value for each group.

The most commonly used SQL aggregate functions are:

MIN() - returns the smallest value of a column
MAX() - returns the largest value of a column
COUNT() - returns the number of rows in a set
SUM() - returns the sum of a numerical column
AVG() - returns the average value of a numerical column
Aggregate functions ignore null values (except for COUNT(*)).

```python
# 이미 있으면 에러가 나므로, 재실행 대비로 try 처리
for col in ["InStock", "InOrder"]:
    try:
        run_execute(f"ALTER TABLE Products ADD COLUMN {col} INTEGER;")
    except Exception as e:
        print(f"{col}: {type(e).__name__} - {e}")
```

```python
import random
random.seed(42)  # 매번 같은 결과가 나오도록 고정

# 모든 ProductID 가져오기
cursor.execute("SELECT ProductID FROM Products;")
ids = [row[0] for row in cursor.fetchall()]

for pid in ids:
    in_stock = random.randint(0, 100)
    # 약 1/3 확률로만 InOrder 에 값, 나머지는 NULL 유지
    in_order = random.randint(0, 50) if random.random() < 0.33 else None
    cursor.execute(
        "UPDATE Products SET InStock = ?, InOrder = ? WHERE ProductID = ?;",
        (in_stock, in_order, pid),
    )
conn.commit()
print("done")
```

```python
query = """
CREATE PROCEDURE procedure_name
  @param1 datatype,
  @param2 datatype
AS
BEGIN
  -- SQL_statements to be executed
  SELECT column1, column2
  FROM table_name
  WHERE columnN = @paramN;
END;

EXEC procedure_name @param1 = 'value1', @param2 = 'value2';
"""
run_select(query)
```
